Linear Algebra(Matrix Operations)

```python
A = np.array([[1, 2],[3, 4]])
B = np.array([[5, 6],[7, 8]])
print(A)
print(B)
```

```python
print(np.dot(A, B))
```

```python
print(np.matmul(A, B))
```

```python
print(np.linalg.inv(A))
```

```python
print(np.linalg.det(A))
```

```python
print(np.linalg.eig(A))
```

```python
print(np.linalg.norm(A))
```

```python
Q, R = np.linalg.qr(A)

print("Q =\n", Q)
print("R =\n", R)
```

```python
U, S, VT = np.linalg.svd(A)

print("U =\n", U)
print("Singular Values =\n", S)
print("V^T =\n", VT)
```

```python
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect("database.db")
cursor = conn.cursor()

# Create employees table
cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    city TEXT,
    salary REAL
)
""")

# Insert sample data
data = [
    (1, "Asha", 25, "Hyderabad", 45000),
    (2, "Ravi", 28, "Chennai", 52000),
    (3, "Kiran", 32, "Bengaluru", 61000),
    (4, "Meena", 29, "Mumbai", 58000),
    (5, "Vikram", 35, "Pune", 72000)
]

cursor.executemany("INSERT INTO employees VALUES (?, ?, ?, ?, ?)", data)
conn.commit()

print(" Table created and data inserted!")
```

```python
print(df.loc[10, "Name"])
```

```python
print(df.loc[5:10, ["Name", "City", "Salary"]])
```
