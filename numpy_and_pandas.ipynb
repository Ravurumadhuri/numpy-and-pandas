Linear Algebra(Matrix Operations)

```python
A = np.array([[1, 2],[3, 4]])
B = np.array([[5, 6],[7, 8]])
print(A)
print(B)
```

```python
print(np.dot(A, B))
```

```python
print(np.matmul(A, B))
```

```python
print(np.linalg.inv(A))
```

```python
print(np.linalg.det(A))
```

```python
print(np.linalg.eig(A))
```

```python
print(np.linalg.norm(A))
```

```python
Q, R = np.linalg.qr(A)

print("Q =\n", Q)
print("R =\n", R)
```

```python
U, S, VT = np.linalg.svd(A)

print("U =\n", U)
print("Singular Values =\n", S)
print("V^T =\n", VT)
```

Random Module(np.random)

```python
a = np.random.rand(3)
b = np.random.rand(2, 3)
print(a)
print(b)

```

```python
x = np.random.randint(10)              # Random int between 0 and 9
y = np.random.randint(5, 20, size=4)   # 4 random ints between 5 and 19
print(x)
print(y)
```

```python
normal_data = np.random.normal(50, 10, 5)
print(normal_data)
```

```python
items = [10, 20, 30, 40, 50]
choice = np.random.choice(items, size=3)
print(choice)
```

```python
arr = np.array([1, 2, 3, 4, 5])
np.random.shuffle(arr)
print(arr)
```

```python
np.random.seed(10)
print(np.random.rand(3))
np.random.seed(10)
print(np.random.rand(3))
```

```python
uniform_data = np.random.uniform(5, 15, size=4)
print(uniform_data)
```

```python
poisson_data = np.random.poisson(lam=4, size=6)
print(poisson_data)
```
